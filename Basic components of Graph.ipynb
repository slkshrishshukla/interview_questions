```python
#3. Setting Labels for X-axis and Y-Axis and Title.
#Setting the Markers Properties
#Draw a plot with the following Values
from matplotlib import pyplot as plt
x=[1,2,3,4]
y=[1,2,3,4]
plt.plot(x,y,ls="solid",lw=6,marker="o",mec="green",mfc="yellow",ms=20,c="red")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Graph")
```
