```python
from  abc import ABC ,abstractmethod 


class Organization(ABC):

    @abstractmethod
    def access_system(self):
        pass
class SecureOrganization(Organization) :
    def __init__(self, user_id,password):
        self.__user_id = user_id
        self.__password = password


    def __authencation(self):
        print("Authenticating user...")
        if self.__user_id == "admin" and self.__password == "secure123":
            return True
        return False
    
    def access_system(self):
        if self.__authencation():
            print('Access Granted !!')
        else:
            print('Access Denied')


org = SecureOrganization("admin", "secure123")
# org = SecureOrganization("admin", "hello")
org.access_system()



```

```python
class Security:
    def __init__(self, user_id, password):
        self.__user_id = user_id  
        self.__password = password  

    def __authenticate(self):  
        print("Authenticating user...")
        return self.__user_id == "admin" and self.__password == "secure123"

    def authenticate_user(self):
        return self.__authenticate() 

class Organization:
    def __init__(self, user_id, password):
        self.security = Security(user_id, password)  

    def access_system(self):
        if self.security.authenticate_user():
            print("Access granted!")
        else:
            print("Access denied!")

# Example usage:
org = Organization("admin", "secure123")
org.access_system()  # Output: Authenticating user... Access granted!
```

Types of Inheritance in Django
1️⃣ Abstract Base Class Inheritance
2️⃣ Multi-Table Inheritance
3️⃣ Proxy Model Inheritance

Used when multiple models need common fields.
✅ The parent model does not create a database table.
✅ Child models inherit the fields but create their own tables.

from django.db import models

class BaseModel(models.Model):  # Abstract Base Class
    created_at = models.DateTimeField(auto_now_add=True)
    updated_at = models.DateTimeField(auto_now=True)

    class Meta:
        abstract = True  # This makes it an abstract model 

class User(BaseModel):  # Inherits BaseModel fields
    name = models.CharField(max_length=100)
    email = models.EmailField(unique=True)

class Product(BaseModel):  # Another model inheriting BaseModel
    name = models.CharField(max_length=100)
    price = models.DecimalField(max_digits=10, decimal_places=2)

----------------------------------------------------------------------------

Multi-Table Inheritance
✅ The parent model creates a database table.
✅ The child model gets its own table with a one-to-one link to the parent.
✅ Used when different child models share a base class but need their own tables.

class Person(models.Model):  # Creates a database table
    name = models.CharField(max_length=100)
    email = models.EmailField(unique=True)

class Employee(Person):  # Creates a separate table
    job_title = models.CharField(max_length=100)


3️⃣ Proxy Model Inheritance
✅ Does not create a new database table.
✅ Used to modify behavior of an existing model (e.g., change ordering).
✅ Useful when you need a customized version of an existing model. 

class UserProfile(models.Model):
    name = models.CharField(max_length=100)
    email = models.EmailField(unique=True)

class AdminUser(UserProfile):  # Creates a separate table
    is_admin = models.BooleanField(default=True)

class OrderedUser(UserProfile):  # Proxy Model (No new table)
    class Meta:
        ordering = ["name"]
        proxy = True  # This makes it a proxy model

```python
# 5. Question: Implement a function to find all permutations of a given string.
nums = ["abc", "acb", "bac", "bca", "cab", "cba"]

def permute( nums):
        result = []
        for i in range(len(nums)):
            char = nums[i]
            remaining_char = nums[:i] + nums[i + 1:]
            for num in permute(remaining_char):
                result.append([char] + num)
        return result
print(permute(nums))

```

```python
def merge_sort(arr):
    

    if len(arr) <= 1:
        return arr 

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    left_sorted = merge_sort(left_half)
    right_sorted = merge_sort(right_half)

    return merge(left_sorted, right_sorted)

def merge(left, right):
    

    merged = []
    left_index = 0
    right_index = 0

    while left_index < len(left) and right_index < len(right):
        if left[left_index] <= right[right_index]:
            merged.append(left[left_index])
            left_index += 1
        else:
            merged.append(right[right_index])
            right_index += 1

    merged.extend(left[left_index:])
    merged.extend(right[right_index:])

    return merged

# Example usage:
my_list = [38, 27, 43, 3, 9, 82, 10]
sorted_list = merge_sort(my_list)
print(sorted_list) 
```
