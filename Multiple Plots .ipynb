```python
from  matplotlib import pyplot as plt 
import numpy as np 
x = [2,1,3,4]
y = [2,3,4,1] 
plt.xlim(0,6) 
plt.ylim(0,6)
plt.xticks([1,2,3,4,5,6])
plt.yticks([1,2,3,4,5,6])
plt.plot(x,y,marker="o",mec="red",mfc="yellow",c="red",ls="solid",lw=3,ms=10)
plt.grid(axis="both",ls="dotted",lw=2,c="green")
plt.xlabel("X-Axis",fontdict={"family":"arial","color":"blue","size":15})
plt.ylabel("Y-Axis",fontdict={"family":"arial","color":"blue","size":15})
plt.title("Line Plot",fontdict={"family":"Times New Roman","color":"red","size":25})
plt.show()
```

```python
#Drawing Sub Plotting
from matplotlib import pyplot as plt 
import numpy as np 
plt.subplot(1,2,1)
plt.plot([1,2,3,4],[4,3,2,1])
plt.xlabel("x -axis")
plt.ylabel("y -axis")
plt.title("XY Plot-1")

# =-----------------------------=
plt.subplot(1,2,2)
plt.plot([1,2,3,4],[1,2,3,4])
plt.xlabel('x- Axis')
plt.ylabel('y- Axis')
plt.title(' XY Plot-2') 
plt.suptitle("Main Plot contains subplots ")
plt.show()


```
