```python
def reverse_parameter(param):
    if isinstance(param, str):
        return ''.join(reversed(param))  # Using reversed() and join() for strings
    elif isinstance(param, int):
        is_negative = param < 0
        reversed_num = int(str(abs(param))[::-1])  # Handle the sign separately
        return -reversed_num if is_negative else reversed_num
    return None

# Example usage
a = reverse_parameter("abc")
b = reverse_parameter(123)
c = reverse_parameter(-123)
print(a,b,c)
```

```python
def find_pairs(lst,taeget):
    seen = set()
    pairs = []
    for num  in lst:
        diff = taeget - num
        if diff in seen:
            pairs.append((diff,num))
        seen.add(num) 
    return pairs 

print(find_pairs([1,2,3,4,5,6],7))
```

```python
def longest_uniqe_substring(s):
    char_map = {}
    left = max_length = 0
    for right, char in enumerate(s):
        if char in char_map:
            left = max(left,char_map[char]+1)
        char_map[char] = right 
        max_length = max(max_length, right-left+1)
    return max_length 
print( longest_uniqe_substring('abcabcbb'))
```

```python
def convert_to_12_hour(time_str):
    from datetime import datetime
    
    # Parse the 24-hour format time string
    time_obj = datetime.strptime(time_str, "%H:%M")
    
    # Convert to 12-hour format with AM/PM
    return time_obj.strftime("%I:%M %p")

# Example usage
time_24 = "20:05"
time_12 = convert_to_12_hour(time_24)
print(time_12)  # Output: 08:05 PM
```
