```python
expression = "1+2+3+[4+5*{6+7+(8+9)}]+((2*"

def is_balanced(expression):
    stack = []
    bracket_pairs = {')': '(', '}': '{', ']': '['}  # Closing to Opening Map

    for char in expression:
        if char in "({[":  # Push opening brackets to stack
            stack.append(char)
        elif char in ")}]":  # Check closing brackets 
            if not stack or stack[-1] != bracket_pairs[char]:
                return False
            stack.pop()
    return not stack

expression = "1+2+3+[4+5*{6+7+(8+9)}]+((2*"

if is_balanced(expression):
    print('Brackets are properly paired')
else:
    print('Brackets are not properly paired')

```

```python
x = [1, 2, 3, ['test', 4, 5, 6, {1: [8, 9], 2: 10}]]

def flatten(data):
    result = []
    for item in data:
        if isinstance(item, list):  # If item is a list, recurse
            result.extend(flatten(item))
        elif isinstance(item, dict):  # If item is a dict, process its values
            for value in item.values():
                result.extend(flatten(value) if isinstance(value, (list, dict)) else [value])
        else:
            result.append(item)  # Append normal values
    return result

print(flatten(x))

```

```python
def fibonacci(n, memo={}):
   
    if n in memo:
        return memo[n] 
    if n <= 1:
        return n  

    # Compute the Fibonacci number and store it in the memo dictionary
    memo[n] = fibonacci(n - 1, memo) + fibonacci(n - 2, memo)
    return memo[n]

# Example usage
if __name__ == "__main__":
    n = 10
    print(f"The {n}th Fibonacci number is: {fibonacci(n)}")
```

```python
movie_ratings = {
    "Inception": [5, 4, 5, 3, 5],
    "The Dark Knight": [5, 5, 4, 5, 4],
    "Interstellar": [4, 4, 4, 5, 4],
    "Tenet": [3, 3, 4, 2, 4],
    "Dunkirk": [5, 4, 5, 7, 5],
}

def calculate_average_rating(ratings):
    return sum(ratings) / len(ratings)
 
average_ratings = {movie: calculate_average_rating(ratings) for movie, ratings in movie_ratings.items()}
 
best_movie = max(average_ratings, key=average_ratings.get)
 
print("Average Ratings for each movie:")
for movie, rating in average_ratings.items():
    print(f"{movie}: {rating}")
 
print(f"\nThe best movie based on average rating is: {best_movie}")
```

```python
def compress_string(input_string):
    compressed = []
    count = 1
    for i in range(1, len(input_string)):
        if input_string[i] == input_string[i - 1]:
            count += 1
        else:
            compressed.append(input_string[i - 1] + str(count))
            count = 1
    compressed.append(input_string[-1] + str(count))
    return ''.join(compressed)
    
input_string = "aasssdddaassdd"
output = compress_string(input_string)
print(output)  


```

```python
def longest_substring(s):
    char_set = set()
    left = 0
    max_length = 0
    longest_substr = ""

    for right in range(len(s)):
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1
        char_set.add(s[right])
        if right - left + 1 > max_length:
            max_length = right - left + 1
            longest_substr = s[left:right + 1]

    return longest_substr

input_string = "abcdabda"
output = longest_substring(input_string)
print(output)  # Output: abcd
```
