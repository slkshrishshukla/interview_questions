```python
import re

def decode_string(input_string):
    matches = re.findall(r'([a-zA-Z])(\d*)', input_string)
    print(matches)
    result = ""
    for char, num in matches:
        result += char * (int(num) if num else 1)
        
    return f"{input_string[0]}{ result}"

input_string = "a2bc5d"
output = decode_string(input_string)
print(output)  # Output: aaabccccccd
```
