```python
expression = "1+2+3+[4+5*{6+7+(8+9)}]+((2*"

def is_balanced(expression):
    stack = []
    bracket_pairs = {')': '(', '}': '{', ']': '['}  # Closing to Opening Map

    for char in expression:
        if char in "({[":  # Push opening brackets to stack
            stack.append(char)
        elif char in ")}]":  # Check closing brackets 
            if not stack or stack[-1] != bracket_pairs[char]:
                return False
            stack.pop()
    return not stack

expression = "1+2+3+[4+5*{6+7+(8+9)}]+((2*"

if is_balanced(expression):
    print('Brackets are properly paired')
else:
    print('Brackets are not properly paired')

```

```python
expression = "1+2+3+[4+5*{6+7+(8+9)}]+((2*"
def check_brackets(expression):
    pairs ={
        ")":"(",
        "}":"{",
        "]":"["
    }
    stack = []

    for char in expression:
        if char  in '({[':
            stack.append(char)
        else :
            if not stack  or  stack[-1] != pairs[char]:
                return False
            stack.pop()
    return  not stack

if check_brackets(expression):
    print('is carect')
else:
    print('is not carect')
```

```python
# l1 = [1,2,3,['test',4,5,6,{1:[8,9],2:10}]]

l1 =[1, [1, [2, 3, 4, [4, 3, 2]],{5: 6}]]
# output = [1, 2, 3, 'test', 4, 5, 6, 8, 9, 10]
 
def remove_list_func(l1):
    result = []
    for char in l1:
        
        if isinstance(char,list):
            
           result.extend(remove_list_func(char))
           
        elif isinstance(char, dict):
             
            for v in char.values():
                result.extend(remove_list_func(v)  if isinstance(v, (list, dict)) else [v])
        else:
            result.append(char)
    return result
    

print(remove_list_func(l1))
                 
```
