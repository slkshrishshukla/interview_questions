```python
def min_distance(S, word1, word2):
    indices_word1 = []
    indices_word2 = []

    for index, word in enumerate(S):
        if word == word1:
            indices_word1.append(index)
        elif word == word2:
            indices_word2.append(index)

    min_dist = float('inf')
    
    for i in indices_word1:
        for j in indices_word2:
            min_dist = min(min_dist, abs(i - j))
    
    return min_dist

# Input data
S = ["the", "quick", "brown", "fox", "quick"]
word1 = "the"
word2 = "fox"

# Output result
result = min_distance(S, word1, word2)
print(result)  # Output: 3
```

```python
def longest_palindromic_substring(s: str) -> str:
    if not s:
        return ""
    
    def expand_around_center(left: int, right: int) -> str:
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1
        # Return the palindrome substring
        return s[left + 1:right] 
    longest = " " 
    for i in  range(len(s)):
        old_p = expand_around_center(i, i)
        if len(old_p) > len(longest):
            longest =old_p 
        eve_p = expand_around_center(i, i+1)
        if len(eve_p) > len(longest):
            longest = eve_p 
    return longest

leg = longest_palindromic_substring('babad')
print(leg)
```

```python
# . Count and Group Anagrams from a List of Strings

from  collections  import defaultdict 

def  group_anagrams(words: list):
    anagram = defaultdict(list)
    for word in words:
        sorted_words = ''.join(sorted(word))
        anagram[sorted_words] .append(word)
    return dict(anagram) 
print(group_anagrams(["eat", "tea", "tan", "ate", "nat", "bat"]))
```

```python
def full_justify(words, maxWidth):
    result = []
    current_line = []
    current_length = 0

    for word in words:
        if current_length + len(word) + len(current_line) > maxWidth:
            for i in range(maxWidth - current_length):
                current_line[i % (len(current_line) - 1 or 1)] += " "
            result.append("".join(current_line))
            current_line = []
            current_length = 0
        current_line.append(word)
        current_length += len(word)
    
    result.append(" ".join(current_line).ljust(maxWidth))
    
    return result

words = ["This", "is", "an", "example", "of", "text", "justification."]
maxWidth = 16

output = full_justify(words, maxWidth)

for line in output:
    print(f'"{line}"')
```

```python
# Find the missing number in a list of consecutive integers.
def find_missing_numbers(l1):
    min_num = min(l1)
    max_num = max(l1)
    full_set = set(range(min_num, max_num + 1))
    input_set = set(l1)
    missing_numbers = sorted(full_set - input_set)
    return missing_numbers

l1 = [1, 2, 4, 7]
print(find_missing_numbers(l1))  
# Output: [3, 5, 6]
```

```python
# . In how many ways you can pay N rupees with 1 rupee, 2 rupee & 5 rupee denominations, In such a way that number of 1 rupee coins
#  always greater than that of 2 rupee coins and number of 2 rupee coins always greater than that of 5 rupee coins. At least one coin 
# should be given from each denomination.



def count_ways(N, x=0, y=0, z=1):
    # Base case: If the sum is exactly N and conditions hold
    if x + 2*y + 5*z == N and x > y > z >= 1:
        return 1
    
    # If the sum exceeds N, return 0
    if x + 2*y + 5*z > N:
        return 0

    # Try increasing the number of coins in a valid order
    count = 0
    if y > z:  # Ensure y > z before increasing x
        count += count_ways(N, x+1, y, z)
    if x > y:  # Ensure x > y before increasing y
        count += count_ways(N, x, y+1, z)
    count += count_ways(N, x, y, z+1)  # Always try increasing z

    return count

N = 10  # Change N to test different amounts
print(count_ways(N))  # Output: Number of ways
```

```python
# Problem: Given a string s containing {}, (), [], determine if it is valid.

def is_valid(s):
    stack = [] 
    mapping ={ ')' : '(', 
                '}': '{',
                ']' :'[' }
    for char in s:
        if char in mapping:
            top_ele =stack.pop() if stack else '#'
            if mapping[char] != top_ele:
                return False 
        else:
            stack.append(char)
    return not stack 

    

```

```python
def max_distinct_char_between_y(X,Y,):
    indices = [i for i ,c  in enumerate(X) if c == Y ]
    if len(indices) < 2:
        return -1 
    
    max_distance = 0 
    for i in range(len(indices)):
        for j in range(i+1, len(indices)):
            sub_str = X[indices[i] +1 : indices[j]]
            filtered_chars = set(c for c in sub_str if c != ' ' and c != Y)
            max_distance = max(max_distance ,len(filtered_chars))
    return max_distance

X =" abc  aabc  bcadab"
Y ="a"
print(max_distinct_char_between_y(X,Y))
```

```python
import pandas as pd

class Person:
    """Represents a person with a name, age, and city."""
    def __init__(self, name, age, city):
        self.name = name
        self.age = age
        self.city = city
        self.age_group = self.categorize_age()

    def categorize_age(self):
        """Classify age into 'minner', 'mager', or 'sonner'."""
        if self.age < 20:
            return "minner"
        elif 20 <= self.age < 40:
            return "mager"
        else:
            return "sonner"

class AgeClassifier:
    """Classifies people into age groups and organizes them."""
    def __init__(self, people):
        self.people = [Person(**p) for p in people]

    def group_by_age(self):
        """Group people by their age category."""
        grouped = {}
        for person in self.people:
            if person.age_group not in grouped:
                grouped[person.age_group] = []
            grouped[person.age_group].append(person.name)
        return grouped

# Sample data
people_data = [
    {"name": "Shrish", "age": 18, "city": "Bangalore"},
    {"name": "Hari", "age": 25, "city": "Mumbai"},
    {"name": "Amit", "age": 42, "city": "Delhi"},
    {"name": "Rahul", "age": 35, "city": "Pune"},
    {"name": "Ravi", "age": 12, "city": "Chennai"},
    {"name": "Vikas", "age": 50, "city": "Kolkata"}
]

# Create an instance of AgeClassifier
classifier = AgeClassifier(people_data)

# Get grouped data
result = classifier.group_by_age()

# Print the result 
print(result)
```

```python
def merge_sort(arr):
    
    if len(arr) <= 1:
        return arr  
    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])  
    right_half = merge_sort(arr[mid:])  

    return merge(left_half, right_half)

def merge(left, right):
   
    sorted_list = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            sorted_list.append(left[i])
            i += 1
        else:
            sorted_list.append(right[j])
            j += 1

    sorted_list.extend(left[i:])
    sorted_list.extend(right[j:])

    return sorted_list

unsorted_list = [38, 27, 43, 3, 9, 82, 10]
sorted_list = merge_sort(unsorted_list)
print("Sorted List:", sorted_list)
      
```
