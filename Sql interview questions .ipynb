```python
-- function 
CREATE FUNCTION get_emp_salary(p_emp_id NUMBER) RETURN  NUMBER AS  v_salary NUMBER;

BEGIN 
    SELECT salary INTO v_salary
    FROM employees
    where employees_id =p_emp_id 
    RETURN v_salary;
END ;

-- Call the function directly in SQL queries.
-- Execute
SELECT get_employee_salary(101) AS salary;


```

```python
create  procedure update_and_return_salary( IN 
        p_emp_id  INT 
        P_increment decimal(10,2)
        P_update_salary  Decimal(10,2)  OUT    -- output 
)
AS 
begin 
    update emp 
    set salary = salary + p_increment
    where emp_id =  p_emp_id

    select P_update_salary =salary
    from emp 
    where emp_id = p_emp_id

end
```

```python
-- function in sql 
CREATE FUNCTION calculate_total_price (
    @quantity INT,
    @unit_price DECIMAL(10,2)
)
RETURNS DECIMAL(10,2)
AS
BEGIN
    RETURN @quantity * @unit_price;  
END;


select calculate_total_price(5, 20.30) as TotalPrice 
```

```python

CREATE FUNCTION get_full_name (
    @first_name NVARCHAR(50), 
    @last_name NVARCHAR(50)
)
RETURNS NVARCHAR(100)
AS
BEGIN
    RETURN @first_name + ' ' + @last_name;  
END;
```
