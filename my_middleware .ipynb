```python
def find_palindrom(s):
    # s = s.strip() 
    # n = len(s)
    palindromes = []

    for i in range(len(s)):
        for j in range(i + 2, len(s) + 1):  
            substr = s[i:j]
            if substr == substr[::-1]:
                palindromes.append(substr)
    
    return palindromes

s = 'pythonmadamprogramdm'
print(find_palindrom(s))
    
    
```
