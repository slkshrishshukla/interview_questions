```python
#Given Problem
#Number of Students in  Class		Heights
#			20						4.5 5.5 6.5 3.5 5.5
#									3.5 5.5 6.2 3.5 5.5
#									3.5 5.5 4.5 3.5 5.5
#									2.5,3.5,6.0,6.0,5.5
#Draw the bar Plot the above Given Data 
lst =[4.5 ,5.5, 6.5 ,3.5, 5.5,3.5, 5.5, 6.2, 3.5 ,5.5,3.5 ,5.5 ,4.5, 3.5, 5.5,2.5,3.5,6.0,6.0,5.5]
d = {}
for val in lst :
    if val not in d:
        d[val] = 1 
    else:
        d[val] = d[val] + 1
print(d)
for  k,v  in d.items():
    print(k ,'  ---->',v)
```

```python
import pandas as pd 
df = pd.DataFrame(d,index=[1])
print(df)
df.columns
```
