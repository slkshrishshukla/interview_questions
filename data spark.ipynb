```python
def binary_search_fun(arr, target):
    low ,high = 0  , len(arr) -1

    while low <= high:
        mid = (low + high) // 2 

        if arr[mid] == target :
            return mid 
        elif  arr[mid] < target :
            low  = mid + 1
        else:
            high =mid - 1 
    return 'not found  arr in list'

arr = [1, 3, 5, 7, 9, 11, 15]
target = 0
print(binary_search_fun(arr,target))
```

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class linkedlist:
    def __init__(self):
        self.head = None 
    def append(self,data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node 
            return 
        temp =self.head 
        while temp.next:
            temp =temp.next
        temp.next =new_node 

    def display(self):
        temp =self.head 
        while temp :
            print(temp.data,end= '  --')
            temp =temp.next 
        print("None")


```
