```python
def upper_func(str1):
    res = str1[::-1]
    res1 = str1.upper()
    
    
    def decorator(func):
        def wrapper(*args, **kwargs):
            result = func(*args, **kwargs)
            return f" {res} -- {res1} "
        return wrapper
    return decorator
    
@upper_func("shrish shukla")
def hello_word():
    return f"hello_word_function"
    
hello_word()
```

```python
def add_two_stirng(str1, str2):
    abc = str1 + ' ' + str2 
    a  = str1.upper() + '   '+ str2.upper()
    res = str1[::-1] + ' '+ str2[::-1]
    
    def decrator_fun(func):
        def wrapper(*args,**kwargs):
            result = func(*args,**kwargs)
            return f' {result} {abc} - {a} --{res}'
        return wrapper
    return decrator_fun

str1 ="shrish" 
str2 = "shukla"
@add_two_stirng(str1, str2)
def hello_word():
    return f"hello_word"

print(hello_word())
```
